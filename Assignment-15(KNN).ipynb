### 2. What are the Parametric and Nonparametric Machine Learning Algorithms

```python
Parametric Machine Learning Algorithms:
    1. Assume a specific functional form for the data distribution.
    2. Have a fixed number of parameters that need to be estimated.
    3. Efficient and require fewer training samples, but rely on strong assumptions.
    4. Examples include Linear regression, Logistic regression, Naive Bayes.

Nonparametric Machine Learning Algorithms:
    1. Do not make explicit assumptions about the data distribution.
    2. Learn the distribution directly from the training data.
    3. More flexible and can capture complex relationships, but may require more training samples and can be computationally 
       expensive.
    4. Examples include Decision trees, Random forest, k-nearest neighbors(KNN), and support vector machines.
```

### 9. What distance metrics can be used in KNN?

```python
Several distance metrics can be used in the K Nearest Neighbor (KNN) algorithm, depending on the nature of the data and 
the specific problem

Some commonly used distance metrics in KNN are:
1. Euclidean Distance: 
    It calculates the straight-line distance between two points in a Euclidean space. It is the most commonly used 
    distance metric in KNN.

2. Manhattan Distance: 
    Also known as the city block distance or L1 norm, it measures the sum of absolute differences between the coordinates 
    of two points.

3. Minkowski Distance: 
    It is a generalization of both the Euclidean and Manhattan distances and is controlled by a parameter 'p'. 
    When 'p' is set to 1, it becomes the Manhattan distance, and when 'p' is set to 2, it becomes the Euclidean distance.
```

```python
Q. What is the formula for Euclidean distance and Manhattan distance?
--> 
Euclidean Distance (ED):
    The Euclidean distance is defined as the shortest path distance between two points source and destination or the straight line distance..
    ED = √[(x2 – x1)² + (y2 – y1)²] (for 2D space)
    ED = sqrt((x1 - y1)² + (x2 - y2)² + ... + (xn - yn)²) (for n-dimensional space)

Manhattan Distance (MD):
    The Manhattan distance between two points is the sum of absolute difference of the coordinates.
    MD = |x2 - x1| + |y2 - y1| (for 2D space)
    MD = |x1 - y1| + |x2 - y2| + ... + |xn - yn| (for n-dimensional space)
```

### 13. What is Normalization?

```python
Normalization is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1. It is also known as Min-Max scaling.
Here’s the formula for normalization:

Normalization equation: X'= X-Xmin/ Xmax-Xmin
  Here, Xmax and Xmin are the maximum and the minimum values of the feature, respectively.
- When the value of X is the minimum value in the column, the numerator will be 0, and hence X’ is 0
- On the other hand, when the value of X is the maximum value in the column, the numerator is equal to the denominator, and thus the value of X’ is 1
- If the value of X is between the minimum and the maximum value, then the value of X’ is between 0 and 1
```

### 14. What is Standardization?

```python
Standardization is another scaling method where the values are centered around the mean with a unit standard deviation. This means that the mean of the attribute becomes zero, and the resultant distribution has a unit standard deviation.
Standardization formula: Z = (X - μ) / σ
Where:
Z is the standardized value of the feature.
X is the original value of the feature.
μ is the mean of the feature.
σ is the standard deviation of the feature.
```

### 16. Why KNN Algorithm is called as Lazy Learner?

```python
KNN is often referred to as a lazy learner. This means that the algorithm does not use the training data points to do any generalizations. In other words, there is no explicit training phase. Lack of generalization means that KNN keeps all the training data. It is a non-parametric learning algorithm because it doesn’t assume anything about the underlying data.
```

### 17. Why should we not use the KNN algorithm for large datasets?

```python
1. The Problem in processing the data:
    KNN works well with smaller datasets because it is a lazy learner. It needs to store all the 
    data and then make a decision only at run time. It includes the computation of distances for 
    a given point with all other points. So if the dataset is large, there will be a lot of 
    processing which may adversely impact the performance of the algorithm.

2. Sensitive to noise:
    Another thing in the context of large datasets is that there is more likely a chance of noise 
    in the dataset which adversely affects the performance of the KNN algorithm since the KNN 
    algorithm is sensitive to the noise present in the dataset.
```

### 18. What are the advantages and disadvantages of the KNN algorithm?

```python
Advantages:
1. No Training Period: It does not learn anything during the training period since it does 
   not find any discriminative function with the help of the training data. In simple words,
   actually, there is no training period for the KNN algorithm. It stores the training 
   dataset and learns from it only when we use the algorithm for making the real-time 
   predictions on the test dataset.
2. As a result, the KNN algorithm is much faster than other algorithms which require training.
   For Example, SupportVector Machines(SVMs), Linear Regression, etc.
3. Moreover, since the KNN algorithm does not require any training before making predictions as a
   result new data can be added seamlessly without impacting the accuracy of the algorithm.
4. Easy to implement and understand: To implement the KNN algorithm, we need only two parameters
   i.e. the value of K and the distance metric(e.g. Euclidean or Manhattan, etc.). Since both the 
   parameters are easily interpretable therefore they are easy to understand.
        
Disadvantages:
1. Does not work well with large datasets: In large datasets, the cost of calculating the distance
   between the new point and each existing point is huge which decreases the performance of the 
   algorithm.
2. Does not work well with high dimensions: KNN algorithms generally do not work well with high 
   dimensional data since, with the increasing number of dimensions, it becomes difficult to 
   calculate the distance for each dimension.
3. Need feature scaling: We need to do feature scaling (standardization and normalization) on the 
   dataset before feeding it to the KNN algorithm otherwise it may generate wrong predictions.
4. Sensitive to Noise and Outliers: KNN is highly sensitive to the noise present in the dataset 
   and requires manual imputation of the missing values along with outliers removal.
```

### 21. How to tune hyperparameters in K Nearest Neighbors Classifier?

```python
1. Define a range of values for hyperparameters, such as K and distance metric.
2. Split the data into training and validation sets.
3. Iterate through hyperparameter combinations, training the model on the training set and evaluating on the validation set.
4. Choose the hyperparameter combination with the best validation performance.
```

### 22. What are the approaches for Hyperparameter tuning?

```python
The approaches for hyperparameter tuning in machine learning include Grid Search, which 
systematically explores all possible combinations of hyperparameters, and Random Search, which 
randomly samples hyperparameters from predefined distributions. Other advanced approaches include 
Bayesian Optimization, Genetic Algorithms, and model-based optimization.
```

### 23. What is cross-validation?

```python
Cross-validation in machine learning is a technique where the available data is divided into 
multiple subsets or folds. The model is trained on a combination of these folds and evaluated on 
the remaining fold. This process is repeated multiple times to assess the model's performance and 
generalize its ability to unseen data.
```

### 27. How do you deal with missing values in a dataset?

```python
KNN imputation is a fairer approach to the Simple Imputation method. It operates by replacing 
missing data with the average mean of the neighbors nearest to it.
```

### 28. How do you use Knn for missing value imputation?

```python
The idea in kNN methods is to identify ‘k’ samples in the dataset that are similar or close in 
the space. Then we use these ‘k’ samples to estimate the value of the missing data points. 
Each sample’s missing values are imputed using the mean value of the ‘k’-neighbors found in the 
dataset.
```

### 29. What is NaN Euclidean Distance?

```python
NaN Euclidean Distance KNN is an extension of the k-nearest neighbors (KNN) algorithm that 
handles missing values in the dataset. It calculates the distance between data points by 
ignoring the missing values during the distance calculation. This modification allows the 
algorithm to select neighbors based on the available features, improving its ability to handle 
datasets with missing values and making more accurate predictions or classifications.
```

### 1. Explain the Decision Tree algorithm in detail.

```python
1. It is supervised Machine learning algorithm which is used to predict categorical as well as
   continuous data.
   i.e classification as well as Regression problem.
2. It is non-parametric algorithm(No assumption on distribution of data).
3. It has a hierarchical, tree structure, which consists of a root node, branch node, internal nodes
   and leaf nodes.
```

### 2. What are the Steps for Making a decision tree?

```python
1. Begin the tree with the root node, says S, which contains the complete dataset.
2. Find the best attribute in the dataset using Attribute Selection Measure (ASM).
3. Divide the S into subsets that contains possible values for the best attributes.
4. Generate the decision tree node, which contains the best attribute.
5. Recursively make new decision trees using the subsets of the dataset created in step -3. 
Continue this process until a stage is reached where you cannot further classify the nodes and 
called the final node as a leaf node.
```

### 4. What are Parametric and Nonparametric Machine Learning Algorithms

```python
Q. Parametric algorithms
-->
1.Parametric algorithms are based on a mathematical model that defines the relationship between 
  inputs and outputs.
2.Parametric algorithms are most appropriate for problems where the input data is well-defined 
  and predictable.

Some examples of parametric algorithms include:
    Linear regression, Logistic regression, Naive Bayes, Simple Neural Networks. 

    
Q. Nonparametric algorithms
-->
1. Nonparametric algorithms are not based on a mathematical model; instead, they learn from the 
   data itself. 
2. Nonparametric algorithms are most appropriate for problems where the input data is not 
   well-defined or is too complex to be modelled using a parametric algorithm.
    
Some examples of Non-parametric algorithms include:
    Decision Trees, k-Nearest Neighbors, Support Vector Machines
```

### 5. Explain Decision tree Key terms: Root Node, Decision Node/Branch Node, Leaf or Terminal Node.

```python
Root Node:
    1. It is the node present at the beginning of a decision tree from this node the population 
       starts dividing according to various features.
    2. Root Node is also called as Decision Node.
    3. Each outcome will lead to additional node.
    
Decision Node/Branch Node: 
    1.The nodes we get after splitting the root nodes are called Decision Node.
    2.When sub-node splits further into another sub-node then it is called decision node.
    
Leaf or Terminal Node:
    1. The nodes where further splitting is not possible are called leaf nodes or terminal nodes.
    2. Leaf node are final node and tree cannot be seggregated further after getting leaf node.
```

### 6. What are Assumptions while creating a Decision Tree?

```python
1. Feature Independence: 
    Decision trees assume that the features used to make decisions are independent of each other.
    This assumption simplifies the modeling process but may not hold in all cases.
2. Homogeneous Data: 
    Decision trees assume that the training data used to build the tree is homogeneous, meaning 
    that the same set of decision rules applies to all instances in the dataset.
3. Attribute Type: 
    Decision trees work well with both categorical and numerical attributes, assuming that the 
    data is appropriately encoded and can be effectively split based on these attribute types.
4. Predictive Accuracy: 
    Decision trees assume that the most predictive features are selected for decision-making. 
    They prioritize features that have a strong correlation with the target variable.
5. No Missing Values: 
    Decision trees typically assume that there are no missing values in the dataset. Some methods 
    can handle missing data, but imputation or preprocessing may be required.
6. No Overfitting: 
    Decision trees can easily overfit the training data if not properly controlled. Techniques 
    like pruning, setting a maximum depth, or using ensemble methods can mitigate overfitting.
```

### 7. What is entropy?

```python
1. Entropy is calculated in a Decision Tree to optimise it. These subsets that complement the 
   Decision Tree features are chosen to achieve greater purity by calculating Entropy. 
   It determines the purity of the component in the subgroup and splits the input accordingly. 
2. It is used to measure purity and impurity of attributes
3. Entropy range lies between 0 to 1.
    If E(s) = 0, Then it is pure node.
    If E(s) > 0, Then it is impure node.
    
Formula:
    E(s) = -P(y)*log2 P(y) -P(N)*log2 P(N)
OR  H(s) =   (pi * log2 pi) 
    Where, Pi is Probability of Target Column.
    Entropy(S) represents the entropy of a set S, which is a subset of the training 
    data at a particular node in the decision tree.
4. Entropy is machine learning metrics that measures the unpredictibility or impurity in system
5. For larger datsets Entropy is not preferble.
```

### 8. What is Information Gain?

```python
1. Information Gain is used in decision tree to decide the best split.
2. Information Gain is basic criterion to decide whether a feature should be used to split a node or not.
3. Information gain is used for determining the best features/attributes that render maximum 
   information about a class. It follows the concept of entropy while aiming at decreasing the 
   level of entropy, beginning from the root node to the leaf nodes.  
4. The feature having the highest value of information gain is accounted for as the best feature 
   to be chosen for split.
    
Information Gain = Entropy before splitting - Entropy after splitting
IG = E(s) = [Weighted Average * Entropy of Each Feature]
```

### 9. What is Gini Index?

```python
1. Gini Index decreases the impurity from the root node to leaf node of decision tree model.
2. Column with low gini index will be considered as root node.
3. For larger datasets GINI is preferable, bcoz there is no logarithmic function.
4. Gini Index range lies between 0 to 0.5.

Formula:
    GI = 1- (pi)**2 
       = 1- P(y)**2 - P(N)**2
        Where, Pi is Probability of Target Column.
```

### 11. What is the difference between Entropy and Gini Impurity(Gini Index)

```python
Entropy:
1. Range 0 to 1
2. -P(Y)log2(P(Y)) - P(N)log2(P(N))
3. For large data set Entropy is not preferable
4. performs slow due to logarithmic function.

Gini Index:
1. Range 0 to 0.5
2. 1-P(Y)**2 - P(N)**2
3. For large data set gini works
4. speed faster than entropy because of there is no logarithmic function.
```

### 12. How do you calculate the entropy of children nodes after the split based on a feature?

```python
1. Split the data based on the feature into subsets.
2. Calculate the entropy of each child node using the proportion of instances in each class.
3. Calculate the weighted average of the entropies based on the proportion of instances in each 
   child node.
It is calculated using the below formula:
Entropy(s) =   (pi * log2 pi) where S is the dataset, Pi is Probability of Target Column.
```

```python
The goal of the feature selection while building a Decision Tree is to select features or 
attributes (Decision nodes) which lead to a split in children nodes whose combined entropy adds 
up to lower entropy than the entropy value of the data segment before the split. This implies 
higher information gain.
```

### 17. What is pruning in the Decision Tree?

```python
1. The branch which are weak or the split which is not giving any information that split is weak 
   split, we cut that branch.
2. Pruning in the Decision Tree is a process of removing unnecessary branches or weak branches to 
   prevent overfitting and improve its generaization ability
    
The process of removing sub-nodes of a Decision node is called pruning, which is the opposite 
process of splitting. 

The two most widely used techniques for pruning are Post and Pre-Pruning.

Post Pruning:

This type of pruning is used after the construction of the Decision Tree.
This technique is used when the Decision Tree has a tremendous depth and will show the overfitting of the model.
It is also known as backward pruning.
This technique is used when we have an infinitely grown Decision Tree.

Pre Pruning:

This technique is used before the construction of the Decision Tree.
Pre-Pruning can be done using Hyperparameter tuning.
Overcome the overfitting issue.
```

### 18. How is a splitting point chosen for continuous variables in decision trees?

```python
1. Sorting the data: Sort the instances in the dataset based on the values of the continuous 
   variable.
2. Evaluating candidate splitting points: Consider each adjacent pair of instances and calculate 
   the midpoint between their values. These midpoints serve as potential splitting points for the 
   continuous variable.
3. Computing the criterion: For each candidate splitting point, calculate the value of the chosen 
   criterion (e.g., information gain, Gini index) for the resulting split.
4. Selecting the best splitting point: Choose the candidate splitting point that maximizes the 
   criterion, indicating the highest information gain or greatest reduction in impurity.
```

### 25. What are the min_samples_split and min_samples_leaf hyperparameters?

```python
min_samples_split:
    The min_samples_split parameter will evaluate the number of samples in the node, and if the 
    number is less than the minimum the split will be avoided and the node will be a leaf.
    By default:min_samples_split=2 
    
min_samples_leaf:
    The min_samples_leaf parameter checks before the node is generated, that is, if the possible 
    split results in a child with fewer samples, the split will be avoided (since the minimum 
    number of samples for the child to be a leaf has not been reached) and the node will be 
    replaced by a leaf.
    By default: min_samples_leaf=1
```

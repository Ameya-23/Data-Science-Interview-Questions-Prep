### 2. What is DBMS and its goals


```python
DBMS stands for Database Management System. It is a software system designed to manage and 
organize large volumes of data efficiently.

The main goals of a DBMS are to provide efficient data management, ensure data integrity and 
security, support data retrieval and manipulation, and enable concurrent access to the database 
by multiple users.
```

### 3. What is RDBMS


```python
- RDBMS stands for Relational Database Management System
- It is structured data in form of tables
- It is a database management system based on a relational model. It facilitates you to store,
  access, and manipulate the data stored in relational databases.
- Relational database stores the data in the collection of tables.
- Examples of relational database management systems are MySQL, SQL Server, Oracle,
  PostgreSQL, etc.
```

### 5. What is SQL and its characteristics


```python
- SQL is a non procedural programming language designed for managing data in relational 
  database management systems (RDBMS).
- SQL was originally developed by IBM for querying, defining and altering relational 
  databases.
- SQL is not case-sensitive. Generally, SQL keywords are written in upper case.
- Using SQL, we can query our database in several ways using English like statements, every 
  SQL statement is called SQL Query.
- SQL follows some unique set of rules and guidelines called syntax.
- Every SQL query ends with a semicolon (;). It is a standard way in a database system in 
  which more than one SQL statement is used in the same call.
- All RDBMS like MySQL, Oracle, SQL Server, Sybase, PostgreSQL, and SQL Server use SQL as 
  a standard database language
```

### 19. Which all changes we might need to perform on an existing database table


```python
Adding a Column: 
    You may need to add a new column to store additional information. This can be done using 
    the "ALTER TABLE" statement with the "ADD" keyword.

Modifying a Column: 
    You might need to modify the data type, length, or constraints of an existing column. 
    This can be done using the "ALTER TABLE" statement with the "ALTER COLUMN" or "MODIFY COLUMN" 
    keywords.

Dropping a Column: 
    If a column is no longer needed, you can remove it from the table using the "ALTER TABLE" 
    statement with the "DROP COLUMN" keyword.

Renaming a Table: 
    If you want to change the name of an existing table, you can use the "sp_rename" command in 
    SQL Server or the "ALTER TABLE" statement with the "RENAME TO" clause in other database 
    systems.

Adding or Modifying Constraints: 
    You may need to add or modify constraints on the table, such as primary keys, foreign keys, 
    unique constraints, or check constraints. This can be done using the "ALTER TABLE" statement 
    with the appropriate keywords.
```

### 26. What is the difference between truncate and drop command


```python
TRUNCATE: 
    It removes all rows from a table, but keeps the table structure intact. 
    It is a data manipulation command.

DROP: 
    It completely removes the table from the database, including the table structure. 
    It is a data definition command.

TRUNCATE allows you to keep the table with the same structure and reinsert data later, 
while DROP requires recreating the table from scratch.        
```

### 1. What is DML and DQL


```python
DML (Data Manipulation Language): 
- DML deals with the manipulation and modification of data stored in a database. 
- It includes statements such as INSERT, UPDATE, DELETE, and MERGE. These statements allow you 
    to add,modify, or remove data within database tables. 
- DML is used to perform operations that change the content of the database.


DQL (Data Query Language): 
- DQL is focused on retrieving and querying data from a database.
- It primarily consists of the SELECT statement, which allows you to specify the columns and 
  conditions to retrieve data from one or more database tables.
- DQL is used to fetch information from the database without modifying its content.
```

### 2. Which commands are part of DML and DQL


```python
DML (Data Manipulation Language):
    INSERT: Adds new records to a table.
    UPDATE: Modifies existing records in a table.
    DELETE: Removes records from a table.
    MERGE: Performs insert, update, or delete operations based on specified conditions.
        
DQL (Data Query Language):
    SELECT: Retrieves data from one or more tables based on specified conditions and columns.
```

### 3. Define CRUD operations and the commands required 


```python
CRUD is an acronym that stands for Create, Read, Update, and Delete. 
These operations represent the four basic functions commonly performed on data in a database or 
any persistent storage system.

Commands:
    
1.Create (C): This operation involves creating new records or entities in a database.
Command: The DML command used to perform the create operation is usually the INSERT statement.
    Ex. INSERT INTO table_name (column1, column2, ...) VALUES (value1, value2, ...);
    
2.Read (R): This operation involves retrieving or querying existing data from a database.
Command: The DQL command used for reading data is the SELECT statement.
    Ex. SELECT column1, column2, ... FROM table_name WHERE condition;

3.Update (U): This operation involves modifying existing data in a database.
Command: The DML command used to perform updates is the UPDATE statement.
    Ex. UPDATE table_name SET column1 = value1, column2 = value2, ... WHERE condition;
    
4.Delete (D): This operation involves removing existing data from a database.
Command: The DML command used for deletion is the DELETE statement.
    Ex. DELETE FROM table_name WHERE condition;
```

### 22. What different types of joins we have in SQL


```python
- INNER JOIN: Returns records that have matching values in both tables.
- LEFT JOIN: Returns all records from the left table and the matched records from the right table.
- RIGHT JOIN: Returns all records from the right table and the matched records from the left table.
- FULL JOIN: Returns all records when there is a match in either the left or right table
```

### 25.Write a SQL query to rank the scores, where rankings should be calculated 
according to the following rules: the scores should be ranked from the highest to 
the lowest; if there is a tie between two scores, both should have the same ranking; 
after a tie, the next ranking number should be the next consecutive integer value 
(i.e., there should be no holes between ranks). Return the result table ordered by 
score in descending order.

(Refer placement table we have for this)

```python
SELECT score,RANK() OVER (ORDER BY score DESC) AS ranking
FROM placement
ORDER BY score DESC;
```

### 26. Write an SQL query to find all numbers that appear at least three times consecutively.



```python
SELECT number
FROM (SELECT number, ROW_NUMBER() OVER (ORDER BY id) AS row_number,
ROW_NUMBER() OVER (PARTITION BY number ORDER BY id) AS sequence_number
FROM your_table) AS subquery 
WHERE row_number - sequence_number >= 2
GROUP BY number
HAVING COUNT(*) >= 3;
```

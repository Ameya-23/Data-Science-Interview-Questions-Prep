### 2. How do you represent a simple linear regression?

```python
1. Simple linear regression is a statistical method for establishing the relationship between two variables using a straight 
   line. The line is drawn by finding the slope and intercept, which define the line and minimize regression errors.
2. It has one input (x) and one output variable (y) and helps us predict the output from trained samples by fitting a 
   straight line between those variables. 
3. In Simple Linear Regression we have only one independent variable and one dependent variable.
    y = mx + c
    where 
    m : Slope
    c : Intercept,the predicted value of y when the x is 0.
    x : Independent Variable
    y : Dependent variable
The above mathematical representation is called a linear equation.
A linear equation is always a straight line when plotted on a graph.

4. In simple linear regression, we assume the slope and intercept to be coefficient and bias, respectively. 
   These act as the parameters that influence the position of the line to be plotted between the data.
```

### 4. What are the assumptions of the Linear regression model?


```python
Linear regression is a useful statistical method we can use to understand the relationship between two variables, x and y. 
However, before we conduct linear regression, we must first make sure that four assumptions are met:
Below are the four assumptions:
1. Linearity: There exists a linear relationship between the independent variable(x) and the dependent variable(y).

2. No multicolinearity: There is a linear relationship between the independent variable(x) and independent variable(x).
                        The residuals are independent. Only in case of independent variable.
    
3. Homoscedasticity: Homoscedasity is the term that states that the spread residuals which we are getting from the 
                     linear regression model should be homogeneous or equal spaces.

4. Normality: The residuals of the model are normally distributed. 
```

### 11. What is VIF? What is the best value of VIF?

```python
Q. What is VIF?
--> 1.Variance inflation factor (VIF) is used to detect the severity of multicollinearity in the ordinary least square (OLS)
      regression analysis.
    2. Multicollinearity inflates the variance and type II error. It makes the coefficient of a variable consistent but 
       unreliable.
    3. VIF measures the number of inflated variances caused by multicollinearity.
Q. What is the best value of VIF?
--> 1. VIF = 1/(1-r2)
     where: r2 = Unadjusted coefficient of determination for regressing the independent variable on the remaining ones 
     In general terms,
     VIF equal to 1 = variables are not correlated
     VIF between 1 and 5 = variables are moderately correlated
     VIF greater than 5 = variables are highly correlated
    2. Where Ri2 represents the unadjusted coefficient of determination for regressing the ith independent variable on the 
       remaining ones. The reciprocal of VIF is known as tolerance. Either VIF or tolerance can be used to detect 
       multicollinearity, depending on personal preference.
```

### 13.What is feature scaling? Is it required in Linear Regression?

```python
Q. What is feature scaling?
-->1. Feature scaling is a technique used in machine learning to standardize the range of features or variables in a dataset.
      It involves transforming the values of different features into a common scale to prevent certain features from 
      dominating others due to their larger magnitude. Feature scaling can help improve the performance of many machine 
      learning algorithms, including linear regression.
    2. It involves transforming the features so that they have a similar scale, typically between 0 and 1 or with a mean of 
       0 and a standard deviation of 1.

Q. Is it required in Linear Regression? 
-->1. In linear regression, feature scaling is not always required, but it can be beneficial in certain cases. 
      The need for feature scaling depends on the specific algorithm being used and the nature of the features in the dataset.
   2. Linear regression models that are based on optimization algorithms like gradient descent can benefit from feature 
      scaling. When features are on different scales, the optimization process may take longer to converge or may even fail 
      to converge. Feature scaling can help speed up convergence and improve the numerical stability of the algorithm.
```

### 19.How to evaluate regression models?

```python
To measure the performance of your regression model,  some statistical metrics are used.
1. Mean Absolute Error(MAE):
    The absolute value turns negative errors into positive ones. MAE=1/N(sum)
    
2. Root Mean Square  Error(RMSE):
    The root mean squared error is simply the square root of the mean squared,RMSE = √sum(ya-yp)**2/N
    
3. Coefficient of determination or R2 or rscore or R-squared: 
    1. It is scale invarient and has fixed range (0 to 1)
    2. R2 determines goodness of BFL. It measures the strength between the dependent and independent variable in range 0 to1 
    3. R-squared= 1-SSE/SST
    
4. Adjusted R2 score:
    Adjusted R2 score value will only increase for correlated features,Formula: 1-(1-R2)*(N-1)/(N-P-1)
        Where, N=Number of data points
               P=Number of predictors or independent variables
               R2=Coefficient of determination

5. Mean Squared Error(MSE):
    The mean squared error takes the sum of squared distances between the model’s predicted values and actual values and 
    divides this sum by the number of test examples.
    Formula: SSE/N = sum(ya-yp)**2/N
```

### 24.What does the coefficient of determination explain?

```python
1. R-squared is a statistical measure of how close the data are to the fitted regression line. 
   It is also known as the coefficient of determination, or the coefficient of multiple determination for multiple 
   regression.

2. The definition of R-squared is fairly straight-forward; it is the percentage of the response variable variation that is 
   explained by a linear model. Or: R-squared = Explained variation / Total variation

3. R-squared is always between 0% and 100%:
    0% indicates that the model explains none of the variability of the response data around its mean. 
    100% indicates that the model explains all the variability of the response data around its mean. 
    In general, the higher the R-squared, the better the model fits your data.
```

### 26.What are the flaws in R-squared?

```python
1. The value of r-square always increases or remains the same as new variables are added to the model, without detecting 
   the significance of this newly added variable (i.e. the value of r-square never decreases on the addition of new 
   attributes to the model. As a result, non-significant attributes can also be added to the model with an increase in the 
   r-square value.
2. The main problem with R-squared is that the R-Square value always increases with an increase in independent variables 
   irrespective of the fact that where the independent variable is contributing to the model or not. 
   This leads to the model having high variance if the model has a lot of independent variables.
```

### 33.How to identify if the model is overfitted or under fitted? Explain in terms of Bias and Variance

```python
Overfitting: 
    
1. A model is overfitted if it has low bias (performs well on the training data) but high variance
   (performs poorly on new data). 
   Example: If training accuracy of the model is 65% and testing accuracy is 50%. then resulted error rate is high because 
            model is unable to capture the relationship between the input and output variables accurately 
            (usually underfit models performs bad for both training and testing data).

   In Optimal Condition, it is said to best/optimal model, when it performs well for training and as well as for testing 
                         dataset.
   Example: if the training accuracy of the model is 85% and testing accuracy of the model is 80–90%. resulting in less 
            error rate. because model is able to capture the relationship between the input and output variables accurately.

Underfitting: 
1. A model is underfitted if it has high bias (performs poorly on both training and new data). 
   It oversimplifies the problem and fails to capture patterns.
   Example: If training accuracy of model is 95% and testing accuracy is 70%. the error rate between training and testing 
            is very high because model is behaving well with training but fails with testing
    

To identify, analyze the model's performance on training and test data. If there's a significant performance gap, 
it indicates overfitting. Consistently low performance suggests underfitting. Striking a balance between bias and variance 
is crucial for good model performance.
```

### 34.How to interpret a Q-Q plot in a Linear regression model?

```python
1. Generate residuals: Fit your linear regression model to the data and obtain the residuals, which are the differences 
                       between the observed values and the predicted values.
2. Sort the residuals: Sort the residuals in ascending order.
3. Compute theoretical quantiles: Calculate the theoretical quantiles for a normal distribution based on the number of 
                                  residuals. These quantiles represent the expected values if the residuals were normally 
                                  distributed.
4. Plot the Q-Q plot: Plot the sorted residuals on the y-axis and the corresponding theoretical quantiles on the x-axis. 
                      Each point on the plot represents a pair of observed and expected values.

5. Assess the plot: Examine the patterns of the points on the Q-Q plot.
                    1. If the points lie approximately along a straight line with a diagonal orientation, it suggests that 
                       the residuals are normally distributed, supporting the assumption of normality.
                    2. If the points deviate significantly from the diagonal line, it indicates departures from normality. 
                       For example:
                        If the points curve upward or downward, it suggests skewness in the residuals.
                        If the points show a distinct S-shape, it indicates heavy-tailed residuals.
```

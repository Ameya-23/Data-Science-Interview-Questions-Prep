### 3. How will you deal with the multiclass classification problem using logistic regression?

```python
1. One-vs-Rest (OVR) Approach: 
    Train multiple binary logistic regression models, each distinguishing one class from the rest.
    During prediction, choose the class with the highest probability among all the classifiers.

2. Multinomial Logistic Regression: 
    Train a single logistic regression model that handles multiple classes directly using the 
    softmax function. Select the class with the highest probability as the predicted class during 
    inference.

Both approaches have their pros and cons, with OVR being simpler but potentially ignoring class 
dependencies, while multinomial logistic regression considers all classes simultaneously. 
Other algorithms like decision trees, random forests, SVMs, or neural networks can also be used 
for multiclass classification depending on the specific problem and data.
```

### 6. What is the formula for the logistic regression function?

```python
The formula for the logistic regression function is as follows:
    p(y)= 1/(1+ e^(-z))
where,
z = mx+c
e = Eulers number = 2.718
y = Target Variable
m = slope
x = Independent Variable
c = intercept

The linear combination mx + c is used as the argument of the exponential function e^(-z), 
where z = mx + c. The exponential term is then used to calculate the probability of the binary 
outcome y being 1. The logistic function ensures that the probability value lies between 0 and 1,
allowing it to model binary classification problems.
```

### 15. What is the Confusion Matrix?

```python
1. The confusion matrix is a NxN matrix used for evaluating the performance of a classification 
   model by providing insights into the types of errors it makes. It is especially useful when 
   dealing with imbalanced datasets, where the number of instances in different classes varies 
   significantly.
2. Matrix compares the actual target values with those predicted values by machine learning model. 
3. The confusion matrix consists of four different metrics:

True Positives (TP): The number of instances correctly predicted as positive by the model.
True Negatives (TN): The number of instances correctly predicted as negative by the model.
False Positives (FP): The number of instances incorrectly predicted as positive by the model 
                      (Type I error).
False Negatives (FN): The number of instances incorrectly predicted as negative by the model 
                      (Type II error).

4. These metrics are used to calculate various performance measures, including accuracy, 
   precision, recall (sensitivity), specificity, and F1 score.
```

### 16. How do you define a classification report?

```python
1.It is one of the performance evaluation metrics of a classification-based machine learning model. 
2.It displays your models precision, recall, F1 score and support. It provides a better 
  understanding of the overall performance of our trained model. 
3.To understand the classification report of a machine learning model, you need to know all of 
  the metrics mentioned below.
 
    1.Precision - Precision is defined as the ratio of true positives to the sum of true and 
                  false positives.Accuracy of positive predictions.
                  Precision = TP/(TP + FP)
    
    2.Recall - Recall is defined as the ratio of true positives to the sum of true positives and 
               false negatives.Fraction of positives that were correctly identified.
               Recall = TP/(TP+FN)
    
    3.F1 Score - The F1 is the weighted harmonic mean of precision and recall. The closer the 
                 value of the F1 score is to 1.0, the better the expected performance of the 
                 model is.
                 F1 Score = 2*(Recall * Precision) / (Recall + Precision)
                
    4.Support - Support is the number of actual occurrences of the class in the dataset. 
                It doesnt vary between models, it just diagnoses the performance evaluation 
                process.
```

### 18. What are the true positive rate (TPR) and true-negative rate (TNR)?

```python
1. True positive rate (TPR):
    1.TPR refers to the ratio of correctly classified positives to the total number of positive 
      instances in the data.
    2.True Positive rate (TRP) is the ratio of True Positives to True Positives and False Negatives. 
      It is the probability that an actual positive will test as positive. 
      TPR = TP / TP + FN
2. True-negative rate (TNR):
    1.TNR refers to the ratio of correctly classified negatives to the total number of negative 
      instances in the data.
      TNR = TN / TN + FP
```

### 19. What is the false-positive rate (FPR) and false-negative rate (FNR)?

```python
1. False-positive rate (FPR):
    1.FPR refers to the ratio of negatives in the data which are incorrectly classified as positives
      to the total number of negative instances in the data.
    2.The False Positive Rate (FPR) is the ratio of the False Positives to all the positives 
      (True positives and false positives). It is actually negative.
      FPR = FP / TP + FP  
2. False-negative rate (FNR):
    1.FNR refers to the ratio of positive instances in the data which are incorrectly classified
      as negatives to the total number of positive instances in the data.
      FNR = FN / TP + FN 
```

### 22. What is the f1 score and Explain its importance?

```python
F1-Score:
    The F1 score is a metric that combines precision and recall into a single value to measure 
    the performance of a classification model. It provides a balanced evaluation, especially in 
    imbalanced datasets.
    F1 = 2 * (precision * recall) / (precision + recall)
    where:
        Precision is the number of true positive predictions divided by the sum of true positive 
        and false positive predictions.
        Recall is the number of true positive predictions divided by the sum of true positive and 
        false negative predictions.
Importance:
    By using the F1 score, one can assess a model's performance on both positive and negative 
    instances, providing a more comprehensive evaluation of its effectiveness. It is especially 
    valuable when the cost of false positives and false negatives differs, as it helps in finding 
    an optimal balance between precision and recall.
```

### 23. Write the equation and calculate the precision and recall rate.

```python
1. To calculate precision and recall rates in machine learning, you need to have a confusion matrix or the actual and 
   predicted labels for a binary classification problem. The confusion matrix typically looks like this:
    
                         Predicted Positive     Predicted Negative
    Actual Positive       TP (True Positive)     FN (False Negative)
    Actual Negative       FP (False Positive)    TN (True Negative)

2. From the confusion matrix, we can define precision and recall as follows:

Precision: 
    It measures the proportion of correctly predicted positive instances out of all instances predicted as positive. 
    It focuses on the accuracy of the positive predictions.
    Precision = TP / (TP + FP)
    
Recall: 
    It measures the proportion of correctly predicted positive instances out of all actual positive instances. It focuses 
    on the coverage of positive instances.
    Recall = TP / (TP + FN)

Example:
    Now, let's assume you have the following confusion matrix for a binary classification problem:
    
                         Predicted Positive     Predicted Negative
    Actual Positive         100 (TP)             20 (FN)
    Actual Negative         10 (FP)              200 (TN)
    
    Using the above values, we can calculate precision and recall rates as follows:
        Precision = TP / (TP + FP) = 100 / (100 + 10) = 0.909 or 90.9%
        Recall = TP / (TP + FN) = 100 / (100 + 20) = 0.833 or 83.3%
        
    So, the precision rate is 90.9% and the recall rate is 83.3% for the given example.
```

### 24. How can you calculate accuracy using a confusion matrix?

```python
To calculate accuracy using a confusion matrix in machine learning, we can use below formula:
    Accuracy = (TP + TN) / (TP + TN + FP + FN)
    where:

    TP (True Positive) is the number of correctly predicted positive values.
    TN (True Negative) is the number of correctly predicted negative values.
    FP (False Positive) is the number of incorrectly predicted positive values.
    FN (False Negative) is the number of incorrectly predicted negative values.
Let's consider the same confusion matrix as before:

                     Predicted Positive     Predicted Negative
    Actual Positive         100 (TP)             20 (FN)
    Actual Negative         10 (FP)              200 (TN)
    
Using the values from the confusion matrix, we can calculate accuracy as follows:
    
    Accuracy = (TP + TN) / (TP + TN + FP + FN) = (100 + 200) / (100 + 200 + 10 + 20) = 0.866 or 86.6%
    
Therefore, the accuracy rate is 86.6% for the given example.
```

### 25. What is sensitivity?

```python
In machine learning, sensitivity is a metric that is also known as recall, hit rate, or true positive rate (TPR). 
It measures the proportion of actual positive instances that are correctly identified as positive by a model.

Sensitivity or recall is calculated using the following formula:
    Sensitivity = TP / (TP + FN) = 1-FNR
    where:
    TP (True Positive) is the number of correctly predicted positive instances.
    FN (False Negative) is the number of incorrectly predicted negative instances.

Sensitivity focuses on the coverage of positive instances and how well a model can identify them correctly. 
A high sensitivity indicates that the model is good at capturing positive instances and has a low false negative rate.
```

### 26. What is Specificity?

```python
In machine learning, specificity is a metric that measures the proportion of actual negative instances that are correctly 
identified as negative by a model. It is also known as true negative rate (TNR).

Specificity is calculated using the following formula:
    Specificity = TN / (TN + FP) = 1-FPR
    where: 
    TN (True Negative) is the number of correctly predicted negative instances.
    FP (False Positive) is the number of incorrectly predicted positive instances.
    
Specificity focuses on the accuracy of identifying negative instances and how well a model can avoid false positives.
```

### 30. What is Overfitting?

```python
1. Overfitting in machine learning refers to a situation where a model performs extremely well on the training data but 
fails to generalize to new, unseen data. It occurs when the model captures the noise or random fluctuations in the training 
data, leading to poor performance on unseen examples.

2. Overfitting can be a challenge in machine learning as it indicates that the model has become too specific to the training
   data and may not perform well in real-world scenarios. Techniques such as regularization, cross-validation, and early 
   stopping are commonly used to mitigate overfitting and improve model generalization.

Reasons for Overfitting are as follows:
    High variance and low bias 
    The model is too complex
    The size of the training data 
    
Techniques to reduce overfitting:
    Increase training data.
    Reduce model complexity.
    Early stopping during the training phase (have an eye over the loss over the training period as soon as loss begins to 
                                              increase stop training).
    Ridge Regularization and Lasso Regularization
    Use dropout for neural networks to tackle overfitting.
```

### 31. What is Underfitting?

```python
A statistical model or a machine learning algorithm is said to have underfitting when it cannot capture the underlying 
trend of the data, i.e., it only performs well on training data but performs poorly on testing data.

Reasons for Underfitting:
    High bias and low variance 
    The size of the training dataset used is not enough.
    The model is too simple.
    Training data is not cleaned and also contains noise in it.

Techniques to reduce underfitting: 
    Increase model complexity
    Increase the number of features, performing feature engineering
    Remove noise from the data.
    Increase the number of epochs or increase the duration of training to get better results.
```

### 32. What is Bias and Variance in Machine Learning?

```python
Q.Bias:
    While making predictions, a difference occurs between prediction values made by the model and actual values/expected 
    values, and this difference is known as bias errors or Errors due to bias.
A model has either:
    Low Bias: A low bias model will make fewer assumptions about the form of the target function.
    High Bias: A model with a high bias makes more assumptions, and the model becomes unable to capture the important 
               features of our dataset. A high bias model also cannot perform well on new data.
Q.Variance:
1. In machine learning, variance refers to the amount of fluctuation or variability in a model's predictions when trained 
    on different subsets of the training data. It measures how much the model's output varies for different values of 
    the training data.
2. High variance indicates that the model is sensitive to the specific examples in the training data and may overfit, 
    while low variance suggests a more stable and generalizable model. Managing variance is important to ensure accurate 
    predictions on unseen data.
```

### 33. What is the tradeoff between Bias and Variance?

```python
While building the machine learning model, it is really important to take care of bias and variance in order to avoid 
overfitting and underfitting in the model. If the model is very simple with fewer parameters, it may have low variance and
high bias. Whereas, if the model has a large number of parameters, it will have high variance and low bias. So, it is 
required to make a balance between bias and variance errors, and this balance between the bias error and variance error is 
known as the Bias-Variance trade-off.
```

### 35. What is L1 and L2 regularization?

```python
1. L1 and L2 regularization are techniques used to prevent overfitting in machine learning models. 
   The regression model that uses L1 regularization technique is called Lasso Regression.
   The regression model that uses L2 regularization technique is called Ridge Regression.

2. L1 regularization (Lasso):
    It adds a penalty proportional to the absolute value of the coefficients, encouraging sparsity and feature selection.
    L1 regularization (Lasso)=  * |w|,
    where  is the regularization parameter and |w| is the sum of the absolute values of the coefficients.
    
3. L2 regularization (Ridge):
    It adds a penalty proportional to the square of the coefficients, encouraging smaller coefficient values to prevent 
    over-reliance on specific features.
    L2 regularization (Ridge)=  * (w^2),
    where  is the regularization parameter and (w^2) is the sum of the squared coefficients.
    
4. Both techniques improve model generalization by adding a penalty term to the loss function, but L1 regularization can 
   set coefficients to zero, while L2 regularization only shrinks them towards zero.
5. Both L1 and L2 regularization aim to reduce the complexity of the model and prevent overfitting by adding a penalty term 
   to the loss function. The amount of regularization is controlled by a hyperparameter called the regularization parameter,
   often denoted as  or alpha.
```
